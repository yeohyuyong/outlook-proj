# Macro Outlook Tracker — Dashboard

Rebuilds `data/entries.csv` from the markdown files in `reports/`, then draws two charts:
1. Forecast trajectories per variable (6 small multiples)
2. Source presence matrix — who's been active, who's gone stale, who's dropped out

```python
import sys
from pathlib import Path
PROJECT_ROOT = Path.cwd().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

import pandas as pd
from src.load import ingest_all
from src.config import ENTRIES_CSV
from src import viz
```

```python
n_runs, n_entries = ingest_all()
print(f'Loaded {n_runs} runs, {n_entries} entries')
```

```python
df = pd.read_csv(ENTRIES_CSV, parse_dates=['run_date', 'source_date'])
df.head()
```

## Viz 1 — Forecast trajectories per variable

One subplot per variable. Each line is one source's forecasts over time; marker size is conviction. If a source gave a directional view but no number, it's skipped here and surfaces instead in Viz 2.

```python
viz.forecast_trajectories(df).show()
```

## Viz 2 — Source presence matrix

Green: present in this run. Yellow: missed this run but last seen within 14 days (stale). Grey: last seen more than 14 days ago — dropped, for now at least.

```python
viz.source_presence_matrix(df).show()
```
